```python
from statsmodels.tsa.stattools import adfuller
def stationarity_check(ts):
    
    
    roll_mean = dfmex['Production in Million of tons Mexico (Mt)'].rolling(center=False,window=12).mean()
    
    plt.plot(ts, color='green',label='Production')
    plt.plot(roll_mean, color='blue', label='Mean Production')
    plt.legend(loc='best')
    plt.title('Produccion Media')
    plt.show(block=False)
    
    #Augmented Dickey-Fuller 
    print('Augmented Dickey-Fuller test:')
    df_test = adfuller(ts)
    print("type of df_test: ",type(df_test))
    print("df_test: ",df_test)
    df_output = pd.Series(df_test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    print("df_output: \n",df_output)
    for key,value in df_test[4].items():
        df_output['Critical Value (%s)'%key] = value
    print(df_output)
```

```python
stationarity_check(dfmex['Production in Million of tons Mexico (Mt)'])
```

```python
import statsmodels.api as sm


fig = plt.figure(figsize=[10, 8])
decompsition = sm.tsa.seasonal_decompose(dfmex['Production in Million of tons Mexico (Mt)'], model='additive')
fig = decompsition.plot()
plt.show()
```

```python
p = d = q = range(0, 2)

p, d, q
```

```python
import itertools
pdq = list(itertools.product(p, d, q))
pdq
```

```python
season_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print(f'Sarimax parameters: {pdq[1]} x {season_pdq[1]}')
```

```python
for param in pdq:
    for param_estacional in season_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(
                dfmex['Production in Million of tons Mexico (Mt)'], 
                order=param,
                seasonal_order=param_estacional, 
                enforce_stationarity=False
            )
            res = mod.fit()
      
            print(f'SARIMAX {param} x {param_estacional}, 12 - AIC: {res.aic}')

        except:
            continue
```
